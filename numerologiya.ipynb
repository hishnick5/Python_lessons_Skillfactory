Всего в жизни придется достигать своим трудом, талантом и умом. Как правило, живут по принципу «хочешь что-то сделать, сделай это сам».  

8-1 Удачливый человек. Человек , имеющий восьмерку никогда не будет обделен материально. Вокруг него всегда складываются благоприятные возможности, удачные стечения обстоятельств. Это люди, которым всегда сопутствует удача, им всегда открыты двери, куда бы не пришли. Успешен в карьере. Обладает деловым мышлением. Компетентный в своей области.
