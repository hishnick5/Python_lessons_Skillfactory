```python
import pandas as pd
import numpy as np
from tabulate import tabulate

input_data = input('Введите дату рождения через точку: День, Месяц, Год ХХ.ХХ.ХХХХ')
# Функция вычисления числа судьбы
def number_destiny(input_data):
    # Проверка на ввод день и месяца в двузначном виде xx.xx.
    def check_month(input_data):
        result = input_data # 
        day, month, year = input_data.split('.') #
        if len(day) == 1: #
            day = '0'+ day #
            result = '.'.join([day, month, year]) #
        if len(month) == 1: #
            month = '0'+ month #
            result = '.'.join([day, month, year]) #
        if len(year) != 4: #
            return 'input Error' #
        return result #
    # Перезаписываем дату вводу
    input_data = check_month(input_data)
    
    # Очищаем дату от нулей и точек
    input_date = input_data.replace('.', '').replace('0', '')
    # Получаем сумму чисел даты рождения в виде строки
    result = str(sum([int(i) for i in input_date]))
    # функция для получения 1-значного числа 
    def calculate_1_number(number):
        # Получаем сумму чисел в виде строки
        result = str(sum([int(i) for i in number]))
        if len(result) > 1: # Если длинна строки больше 1
            # Вызываем функцию рекурсивно в виде аргумента(результат)
            return calculate_1_number(result)
        # Возвращаем полученный результат
        return result
    # Проверяем, если число судьбы не 1 значное
    if len(result) > 1: # Доводим до однозначного с помощью нашей функции
        result = calculate_1_number(result)
    return result

destiny = number_destiny(input_data)

# Функция корректировки 22 чисел
def edit_numb(numb):
    if numb <= 22:
        return numb
    else:
        numb -= 22
    return edit_numb(numb)

# Функция вычисления чисел кармических путей
def karmic_way(input_data):
    if type(input_data) is str:
        input_data = input_data.split('.')
    a = int(input_data[0])
    if a > 22:
        a = edit_numb(a)
    b = int(input_data[1])
    c = sum([int(i) for i in input_data[2]])
    if c > 22:
        c = edit_numb(c)
    d = a+b+c
    if d > 22:
        d = edit_numb(d)
    return a,b,c,d

# Кармический путь по ячейкам
a,b,c,d = karmic_way(input_data)

# ЧД Числа Достижений
ch_d1 = edit_numb(abs(a+b))
ch_d2 = edit_numb(abs(a+c))
ch_d3 = edit_numb(abs(ch_d1+ch_d2))
ch_d4 = edit_numb(abs(b+c))

# КУ Кармические Узлы(уроки)
karmic1 = edit_numb(abs(a-b))
karmic2 = edit_numb(abs(a-c))
karmic3 = edit_numb(abs(karmic1-karmic2))
karmic4 = edit_numb(abs(b-c))
karmic5 = edit_numb(abs(karmic1 + karmic2 + karmic3 + karmic4))

period1 = 36-int(destiny)
period2 = period1 + 9
period3 = period2 + 9
period4 = period3
heads = ['Путь', 'Число Достижений', 'Кармический Узел', 'Период']

karmic_lists = [
                [f'Супер сила ({a})', f'ЧД 1 ({ch_d1})', f'КУ 1 ({karmic1})', f'0 - {period1}'],
                [f'Задача на жизнь ({b})', f'ЧД 2 ({ch_d2})', f'КУ 2 ({karmic2})', f'{period1} - {period2}'],
                [f'Энергия года ({c})', f'ЧД 3 ({ch_d3})', f'КУ 3 ({karmic3})', f'{period2} - {period3}'],
                [f'Предназначение ({d})', f'ЧД 4 ({ch_d4})', f'КУ 4 ({karmic4})', f'{period3} - ∞'],
                [None, None, f'КУ 5 ({karmic5})', f'Число судьбы ({destiny})']
                ]

print(tabulate(karmic_lists, headers=heads, tablefmt="grid", stralign='center', showindex=False))
```
